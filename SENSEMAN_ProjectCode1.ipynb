# Part A(ii)
Creates a function that takes a 6 digit input seed, squares it, removes the side digits and extracts the middle 6 digits as the next input seed.

```python
print('Part A(ii): creating Neumann random arrays...')
def neumann_create(seed):
    neumann = []
    for i in range(1000):
        squared = seed**2
        step1 = squared%1000
        step2 = squared - step1
        step3 = step2//1000
        final = step3%1e6
        neumann.append(final)
        i+=1
        seed = final
    array = np.array(neumann)/1.0e6
    return array
neumann_array_1 = neumann_create(178903)
```

```python
neumann_array_2 = neumann_create(420696)
```

```python
neumann_array_3 = neumann_create(684200)
```

# Part A(iii)
Creates a dictionary containing the Neumann random values as keys, and how many of each value as the items. Then it returns the keys that have values larger than 10, showing that the sequence failed to execute being random.

```python
print('Part A(iii): testing for repeating Neumann arrays...')
def repeat_finder(data):
    unique, counts = np.unique(data, return_counts=True)
    iterations = dict(zip(unique, counts))
    repeating_dict = {}
    for key, value in iterations.items():
        if value > 10:
            repeating_dict[key] = value
    return repeating_dict
print('The following values are repeated more than 10 iterations in Neumann test 1:',repeat_finder(neumann_array_1))
```

# Part B(iii,a,i)
Imports the ChiSquared/pvalue test, but evaluates them in the next section.

```python
print('Part B(iii,a,ii): Creating Chi-Squared tests and P-Values...')
import scipy.stats as stats
def mychi(data,expect):     
    chisq=np.sum((data-expect)**2/expect ) 
    pval=1.0-stats.chi2.cdf(chisq,np.size(data)-1) 
    return chisq,pval
```

# Part B(iii,b,i)
Creates 3 functions to find counts of neighboring data being in the same bin. Then, uses another function to find counts of neighboring data to be in neighboring bins.

```python
print('Part B(iii,b,i): Finding counts of neighboring values in same/neighbor bins..')
pause()
def samebin_random(bin):
    count = 0
    for i in range(2999):
        if (random_data[i] >= bin-.1) and (random_data[i] < bin) and (random_data[i+1] < bin):
            count +=1  
    return count
sum_1 = []
for i in range(10):
    i/10
    sum_1.append(samebin_random(i))
total_1 = sum(sum_1)
print('There are',total_1,'counts of same bin neighboring values for the np.random data.')

def neighborbin_random(dataset):
    count = 0
    for i in range(2999):
        if (np.around(dataset[i],1) == np.around(dataset[i+1],1) - 0.1) or (np.around(dataset[i],1) == np.around(dataset[i+1],1) + 0.1):
            count +=1
    return count
print('There are',neighborbin_random(random_data),'counts of neighbor bin neighboring values for the np.random data.')
```

```python
def samebin_neumann(bin):
    count = 0
    for i in range(2999):
        if (neumann_data[i] >= bin-.1) and (neumann_data[i] < bin) and (neumann_data[i+1] < bin):
            count +=1  
    return count
sum_2 = []
for i in range(10):
    i/10
    sum_2.append(samebin_neumann(i))
total_2 = sum(sum_2)
print('There are',total_2,'counts of same bin neighboring values for the Neumann data.')

print('There are',neighborbin_random(neumann_data),'counts of neighbor bin neighboring values for the Neumann data.')
```

```python
def samebin_human(bin):
    count = 0
    for i in range(2999):
        if (human_data[i] >= bin-.1) and (human_data[i] < bin) and (human_data[i+1] < bin):
            count +=1  
    return count
sum_3 = []
for i in range(10):
    i/10
    sum_3.append(samebin_human(i))
total_3 = sum(sum_3)
print('There are',total_3,'counts of same bin neighboring values for the human.npz data.')

print('There are',neighborbin_random(human_data),'counts of neighbor bin neighboring values for the human.npz data.')
```

1.) A sequence of numbers is said to be pseudorandom because a true random sequence is diifcult to replicate on a machine that can only follow programmed instructions. Pseudorandomness or "fake" randomness uses formulas to generate a sequence of numbers that appear random, and have the properties of true random numbers. I would hope that a pseudorandom sequence has properties like being deterministic and lacks any noticable pattern or repetition. A sequence that's deterministic is useful as that means it can be replicated. Pseudorandom sequences start with a seed number, which kicks off the sequence and using the same seed will produce the same number sequence. Visually, having a sequence that lacks a pattern and repetition seems like a truly random sequence. The numpy.random sequences exhibit both of these traits, particularly the former because of the manual seed input.
